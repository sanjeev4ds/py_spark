```python
from pyspark.sql import SparkSession
spark= SparkSession.builder.appName("spark_3").getOrCreate()

from pyspark.sql.types import StructType, StructField, StringType, IntegerType,DateType

#employee
employees_schema = StructType([
    StructField("emp_id",IntegerType(),False),
    StructField("full_name",StringType(),False),
    StructField("dept_id",IntegerType(),False)
])

employees = [
    (101,"Sanjeev Saini",1),
    (102,"Abhishek Patel",2),
    (103,"Tammanna",3),
    (104,"Anand",4),
    (105,"Akshat",1)
]

emp_df = spark.createDataFrame(data=employees, schema= employees_schema)
print("emp_df")
emp_df.show()

#department
departments_schema = StructType([
    StructField("dept_id",IntegerType(), False),
    StructField("dept_name",StringType(),False),
    StructField("dept_open_date",DateType(),False)
])

from datetime import datetime
departments = [
    (1,"Daas",datetime.strptime("2019-04-01","%Y-%m-%d")),
    (2,"FP",datetime.strptime("2024-01-01","%Y-%m-%d")),
    (3,"Toggle Migration",datetime.strptime("2023-08-01","%Y-%m-%d")),
    (4,"TWD",datetime.strptime("2023-01-01","%Y-%m-%d")),
]
dept_df = spark.createDataFrame(data=departments, schema=departments_schema)
print("dept_df")
dept_df.show()

#current_salaries
current_salaries_schema = StructType([
    StructField('emp_id',IntegerType(),False),
    StructField('salary_amount',IntegerType(), False)
])

current_salaries = [
    (101,15000),
    (102,20000),
    (103,30000),
    (104,40000),
    (105,50000)
]

current_salaries_df = spark.createDataFrame(data=current_salaries, schema=current_salaries_schema)
print("current_salaries_df")
current_salaries_df.show()

#bonuses
bonuses_schema = StructType([
    StructField("bonus_date", DateType(), False),
    StructField("emp_id", IntegerType(), False),
    StructField("amount", IntegerType(), False)
])

bonuses = [
    (datetime.strptime("2024-06-29","%Y-%m-%d"),101,2000),
    (datetime.strptime("2024-06-29","%Y-%m-%d"),102,4000),
    (datetime.strptime("2024-06-29","%Y-%m-%d"),103,6000),
    (datetime.strptime("2024-06-29","%Y-%m-%d"),104,8000),
    (datetime.strptime("2024-06-29","%Y-%m-%d"),105,10000),
    (datetime.strptime("2024-07-30","%Y-%m-%d"),101,8000),
    (datetime.strptime("2024-07-30","%Y-%m-%d"),102,6000),
    (datetime.strptime("2024-07-30","%Y-%m-%d"),103,4000),
    (datetime.strptime("2024-07-30","%Y-%m-%d"),104,2000),
    (datetime.strptime("2024-07-30","%Y-%m-%d"),105,1000),
    (datetime.strptime("2024-08-30","%Y-%m-%d"),101,2000),
    (datetime.strptime("2024-08-30","%Y-%m-%d"),102,4000),
    (datetime.strptime("2024-08-30","%Y-%m-%d"),103,6000),
    (datetime.strptime("2024-08-30","%Y-%m-%d"),104,8000),
    (datetime.strptime("2024-08-30","%Y-%m-%d"),105,10000)
]

bonuses_df= spark.createDataFrame(data=bonuses, schema=bonuses_schema)
print("bonuses_df")
bonuses_df.show()

#salary txn
salaries_txn_schema = StructType([
    StructField("txn_id",IntegerType(), False),
    StructField("emp_id", IntegerType(), False),
    StructField("salary_credit_date", DateType(), False),
    StructField("salary_credit_amount", IntegerType(), False)
])

salaries_txn = [
    (1,101,datetime.strptime("2024-06-29","%Y-%m-%d"),17000),
    (2,102,datetime.strptime("2024-06-29","%Y-%m-%d"),24000),
    (3,103,datetime.strptime("2024-06-29","%Y-%m-%d"),36000),
    (4,104,datetime.strptime("2024-06-29","%Y-%m-%d"),48000),
    (5,101,datetime.strptime("2024-07-30","%Y-%m-%d"),23000),
    (6,102,datetime.strptime("2024-07-30","%Y-%m-%d"),26000),
    (7,103,datetime.strptime("2024-07-30","%Y-%m-%d"),34000),
    (8,104,datetime.strptime("2024-07-30","%Y-%m-%d"),42000),
    (9,101,datetime.strptime("2024-08-30","%Y-%m-%d"),17000),
    (10,102,datetime.strptime("2024-08-30","%Y-%m-%d"),24000),
    (11,103,datetime.strptime("2024-08-30","%Y-%m-%d"),36000),
    (12,104,datetime.strptime("2024-08-30","%Y-%m-%d"),42000),
    (13,105,datetime.strptime("2024-06-29","%Y-%m-%d"),60000),
    (14,105,datetime.strptime("2024-07-30","%Y-%m-%d"),52000),
    (15,105,datetime.strptime("2024-08-30","%Y-%m-%d"),58000)
]

salaries_txn_df = spark.createDataFrame(data=salaries_txn, schema=salaries_txn_schema)
print("salaries_txn_df")
salaries_txn_df.show()
```

```python
from pyspark.sql.functions import *

exact_bonuses_month_df= bonuses_df.withColumn("month",month(col("bonus_date"))).select("emp_id","month","amount")

salary_needs_to_be_credited_df= exact_bonuses_month_df.join(
    current_salaries_df,
    exact_bonuses_month_df.emp_id==current_salaries_df.emp_id,
    how="inner"
    ) \
    .select(exact_bonuses_month_df.emp_id,
            exact_bonuses_month_df.month,
            (exact_bonuses_month_df.amount+current_salaries_df.salary_amount).alias("salary_needs_to_be_credited")
    )
print("salary_needs_to_be_credited_df")
salary_needs_to_be_credited_df.show()

month_salaries_txn_df= salaries_txn_df.withColumn("month",month(col("salary_credit_date"))).select("emp_id","month","salary_credit_amount")
print("month_salaries_txn_df")
month_salaries_txn_df.show()

#original credited- month_salaries_txn_df
#original value needs to be credited- salary_needs_to_be_credited_df

combine_credit_vs_need_df = month_salaries_txn_df.join(
    salary_needs_to_be_credited_df,
    (
        (month_salaries_txn_df.emp_id== salary_needs_to_be_credited_df.emp_id)
        &
        (month_salaries_txn_df.month== salary_needs_to_be_credited_df.month)
    ),
    how="inner"
).select(month_salaries_txn_df.emp_id, month_salaries_txn_df.month, month_salaries_txn_df.salary_credit_amount,salary_needs_to_be_credited_df.salary_needs_to_be_credited) \
.orderBy("month","emp_id")
print("combine_credit_vs_need_df")
combine_credit_vs_need_df.show()

#find gaps
gap_between_credits_txn_df= combine_credit_vs_need_df.where(col("salary_credit_amount")!=col("salary_needs_to_be_credited"))\
.select("emp_id",
        "month",
        "salary_credit_amount",
        "salary_needs_to_be_credited",
         (col("salary_needs_to_be_credited")-col("salary_credit_amount")).alias("gap_amount")
)
print("gap_between_credits_txn_df")
gap_between_credits_txn_df.show()
```

```python
emp_sum_salary_df =salaries_txn_df \
.groupBy("emp_id") \
.agg(sum(col("salary_credit_amount")).alias("total_salary"))
print("emp_sum_salary_df")
emp_sum_salary_df.show()

dept_salary_df= emp_sum_salary_df.join(
    emp_df,
    emp_sum_salary_df.emp_id==  emp_df.emp_id,
    how="inner"
) \
.select("dept_id","total_salary") \
.groupBy("dept_id").agg(sum("total_salary").alias("total_dept_salary")) \
.join(dept_df, "dept_id",how="inner")\
.select("dept_name","total_dept_salary") \
.orderBy(col("total_dept_salary").desc())

print("dept_salary_df")
dept_salary_df.show()
```
